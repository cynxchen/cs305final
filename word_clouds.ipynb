# Creating Lyric Word Clouds

### Load data and prepare

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
# data cleaning and preparation
songs = pd.read_csv('billboard.csv', encoding='latin1')
songs = songs[(songs['Lyrics'].notnull()) & (songs['Lyrics'] != '  ')]
songs['word_count'] = songs['Lyrics'].apply(lambda s: len(s.split()))
songs['Decade'] = songs['Year'].apply(lambda y: str(y//10) + "0's")
songs = songs[songs['word_count'] > 1]
songs.shape
```

```python
songs.head()
```

```python
plt.plot(songs['Year'], songs['word_count'])
plt.show()
```

### Create Word Clouds by Year

```python
def create_cloud_yr(year):
    subset = songs[songs['Year'] == yr]
    text = ' '.join(subset['Lyrics'])
    stopwords = set(STOPWORDS)
    # Create and generate a word cloud image:
    wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
    # Display the generated image:
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(str(year))
    plt.axis("off")
    plt.savefig("clouds/year/" + str(year) + ".png", format="png")
    
```

```python
for yr in range(1965,2016):
    print(yr)
    create_cloud_yr(yr)
```

### Create Word Clouds by Decade

```python
def create_cloud_decade(decade):
    subset = songs[songs['Decade'] == decade]
    text = ' '.join(subset['Lyrics'])
    stopwords = set(STOPWORDS)
    # Create and generate a word cloud image:
    wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
    # Display the generated image:
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(decade)
    plt.axis("off")
    plt.savefig("clouds/decade/" + decade + ".png", format="png")
    plt.show()
```

```python
decades = ["1960's","1970's","1980's","1990's","2000's","2010's",]
for d in decades:
    print(d)
    create_cloud_decade(d)
```
