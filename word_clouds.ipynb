```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
songs = pd.read_csv('billboard.csv', encoding='latin1')
```

```python
songs.head()
songs.shape
```

```python
songs = songs[(songs['Lyrics'].notnull()) & (songs['Lyrics'] != '  ')]
songs.shape
```

```python
songs['word_count'] = songs['Lyrics'].apply(lambda s: len(s.split()))
songs = songs[songs['word_count'] > 1]
songs.shape
```

```python
songs.head()
```

```python
plt.plot(songs['Year'], songs['word_count'])
plt.show()
```

```python
# Start with one review:
text = songs.Lyrics[0]

def create_cloud(year):
    subset = songs[songs['Year'] == yr]
    text = ' '.join(subset['Lyrics'])
    stopwords = set(STOPWORDS)
    # Create and generate a word cloud image:
    wordcloud = WordCloud(stopwords=stopwords, background_color="white").generate(text)
    # Display the generated image:
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.title(str(year))
    plt.axis("off")
    plt.savefig("clouds/" + str(year) + ".png", format="png")
#     plt.show()
    
```

```python
for yr in range(1965,2016):
    print(yr)
    create_cloud(yr)
    
```
